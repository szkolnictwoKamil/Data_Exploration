# Analiza zbioru danych New YorkTimes

#### Zad 1 

*Załaduj zbiór danych NYT jako DataFrame*

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('nyt-frame.csv', header = 0, index_col = 0)
```

```python
df
```

#### Zad 2

*Wyświetl 20 losowych wybranych elementów słownika*

```python
data = df.iloc[:,9:]
array = data.values
```

```python
header = list(df.columns.values[9:])
sample_word = np.random.choice(header, 20, replace=False)
print(sample_word)
```

#### Zad 3

*Dokonaj  transformacji  PCA  pomijając  kolumny,  które  nie  reprezentują słów a jedynie metadane*

```python
pca = PCA()
X_pca = pca.fit_transform(array)
```

#### Zad 4

*Dla pierwszego komponentu wiodącego wyświetl 15 elementów o największej wartości*

```python
df_pca = pd.DataFrame(pca.components_, columns = header)
df_pca.iloc[0].sort_values(ascending = False)[:15:]
```

Najmniejsze wartości pierwszego komponentu odpowiadają takim słowom jak: "music", "trio", "theater" itd.

#### Zad 7

*Dokonaj wizualizacji wektorów cech zrzutowanych na 2(3) pierwsze komponenty wiodące. Zaznacz dwie klasy (art, music) oddzielnymi kolorami*

```python
reds = (df.loc[:, 'class.labels'] == 'art')
blues = (df.loc[:, 'class.labels'] == 'music')

plt.figure()
plt.scatter(X_pca[np.array(reds), 0], X_pca[np.array(reds), 1], c="red")
plt.scatter(X_pca[np.array(blues), 0], X_pca[np.array(blues), 1], c="blue")
plt.title("Projection by PCA")
plt.xlabel("1st component")
plt.ylabel("2nd component")
plt.show()
```

#### Zad 8

*Narysuj wykres zależności wartości wariancji od numeru kierunku wiodącego k.*

```python
variance_ratio = pca.explained_variance_ratio_
plt.plot(variance_ratio, 'ro')
plt.show()
print(f'''Część wariancji, jaka zostanie zachowana po wykonaniu projekcji 
      na pierwsze 10 komponentów wiodących: {sum(variance_ratio[0:10])}''')
```

*Jaka część wariancji zostaje zachowana po wykonaniu projekcji na pierwsze 10 komponentów wiodących?*

Po wykonaniu projekcji na pierwsze 10 komponentów wiodących pozostaje zachowane ok 16% wariancji całkowitej.

# Analiza zbioru danych 04Cars

#### Zad 1

*Załaduj zbiór danych 04Cars jako DataFrame(biblioteka Pandas) i odfiltruj 11 ostatnich kolumn - zostaną one użyte jako wejście dla PCA.*

```python
df_cars = pd.read_csv('04cars-data.csv', header = 0)

df_cars
```

```python
data_cars = df_cars.iloc[:,9:]
array_cars = data_cars.values
array_cars
```

#### Zad 2

*Znormalizuj dane, zapewniając, że dla każdej cechy średnia arytmetyczna wynosi 0, a wariancja 1.*

```python
array_cars_standarized = preprocessing.StandardScaler().fit_transform(array_cars)
array_cars_standarized.mean(axis = 0)
```

```python
array_cars_standarized.std(axis = 0)
```

#### Zad 3

*Dokonaj transformacji PCA dla wejściowego zbioru danych, a następnie przedstaw zależność bezwzględnej i względnej wartości wariancji od numeru kierunku wiodącego k.*

```python
pca_c = PCA()
pca_cars = pca_c.fit_transform(array_cars_standarized)
variance = pca_c.explained_variance_
plt.title("Projection by variance")
plt.plot(variance, 'ro')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
variance_ratio = pca_c.explained_variance_ratio_
plt.title("Projection by variance")
plt.plot(variance_ratio, 'ro')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
print(sum(variance_ratio[0:2]))
print(sum(variance_ratio[0:3]))
```

*Jaka część wariancji zostaje zachowana przy redukcji wymiarowości do 2 i 3*

Przy redukcji wymiarowości do 2 zostaje zachowane ok. 80% wariancji, a przy redukcji do 3 ok. 90% wariancji.

#### Zad 4

*Wyświetl wartości elementów pierwszego i drugiego komponentu wiodącego wraz z odpowiadającą im nazwą cechy.*

```python
header1 = list(df_cars.columns.values[-11:])
attributes1 = df_cars.columns.values[-11:]
components_1 = pca_c.components_[0]
components_2 = pca_c.components_[1]
print('Attribute: \t\tPC1: \t\tPC2:')
for i in range(0, 11):
    print(attributes1[i] + ':\t' + repr(components_1[i]) + ':\t' + repr(components_2[i]))
```
